#variable length argument
(we give any number of arguments)<br>
#types
1) non keyword variable length argument <br>
it behaves like tuple and is represented by star (*)<br>
2) keyword variable length argument <br>
it behaves like dictionary and is represented by double stars (**)

```python
#non keyword variable length argument
def test(*args):
  print(args)
  print(len(args))
  print(type(args))
```

```python
test(1,2,3,4,5,6)
```

```python
def sum_number(*args):
  sum=0
  for num in args:
    sum+=num
  return sum
```

```python
sum_number(1,2,3,4,5)
```

```python
#keyword variable length argument
def test(**kwargs):
  print(kwargs)
  print(len(kwargs))
  print(type(kwargs))
```

```python
test(a=10,b=20,c=30)
```

```python
import numpy as np
from numpy.linalg import norm

A = np.array([2,1,2,3,2,9])
B = np.array([3,4,2,4,5,5])

#compute cosine similarity
cosine = np.dot(A,B) / (norm(A)*norm(B))
print("cosine similarity:",cosine)
```
